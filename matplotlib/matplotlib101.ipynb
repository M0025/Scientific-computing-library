### 2.Draw histogram

```python
x1 = [1,3,5,7,9]
y1 = [5,2,7,8,2]

plt.bar(x1, y1, label='Bar-01')
x2 = [2,4,6,8,10]
y2 = [8,6,2,5,6]
plt.xlabel('X')
plt.ylabel('Y')
plt.bar(x2, y2, label='Bar-02', color='g', width=0.5)
# argument color could use rgb color，logogram also right 

plt.legend()
plt.title('Matplotlib TO draw a histogram')
```

### 3. Bar diagram

```python
x1 = [1,3,5,7,9]
y1 = [5,2,7,8,2]

plt.barh(x1, y1, label='Bar-01')
x2 = [2,4,6,8,10]
y2 = [8,6,2,5,6]
plt.xlabel('X')
plt.ylabel('Y')
plt.barh(x2, y2, label='Bar-02', color='g')
# argument color could use rgb color，logogram also right 

plt.legend()
plt.title('Matplotlib TO draw a bar chart')
```

### Draw frequency distribution histogram 

```python
# Data need to be know frequency distribution
population_ages =[22,55,62,45,21,22,34,42,42,4,99,102,110,120,121,122,130,111,115,112,80,75,65,54,44,43,42,48]
# Divide the data sample classification interval
bins = [0,15,25,40,60,130]
# Use the frequency distribution histogram to count the number of samples within each interval
plt.hist(population_ages, bins)
plt.xlabel("People amounts")
plt.ylabel("Age range")
plt.title("Age frequency distribution bar chart")
```

### Draw sector diagram

```python
players = [20, 12, 7, 2]  # player's amount
types = ['teenger', 'youngth','mid','old']  # player age range
# use datas above to draw a sector diagram
plt.pie(players, 
       labels = types,
       autopct = "%1.1f%%",  # show percent
       startangle = 90  )# anticlockwise
```

### Scatter diagram

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

# label -> lenged; s -> pointsize; marker -> pointStyle
plt.scatter(x, y, label='skitsdcat', color='r', s=25, marker ='*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
```

### Stack maps

```python
days = [1,2,3,4,5]  # public x-axis; time line

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.stackplot(days, sleeping,eating,working,playing,
             colors =['g','c','r','y'],
             labels = ['sleeping','eating','working','playing'])
plt.xlabel('Weekdays')
plt.ylabel('Every activity costs')

# # Draw 4 plots with no data
# plt.plot([],[],color = 'm',label='sleep',linewidth=5)
# plt.plot([],[],color = 'c',label='eating',linewidth=5)
# plt.plot([],[],color = 'r',label='working',linewidth=5)
# plt.plot([],[],color = 'k',label='playing',linewidth=5)
plt.legend()
plt.title('StackMap')
```
