```python
# Get a piece of white paper(dashboard)
fig = plt.figure(1)
# divide dashboard into 1 row 1 col as a coordinate system
ax1 = plt.subplot(111)
# perpare data
data = np.array([15,20,18,25])
# difine width of the bars
width = 0.5
# give the ticks of x-axis
x_bar = np.arange(4)

# Use object ax1 to draw the bars in the coordinate system
rect = ax1.bar(x = x_bar, height = data, color='lightblue')
# Edit ticks of x-axis
ax1.set_xticks(x_bar)
ax1.set_xticklabels(['FirstQuarter','SecondQuarter','ThirdQuarter','FouthQuarter'])
# show grids
ax1.grid(True)
ax1.set_ylabel('Sales/wj')
ax1.set_title('2017 quarters sales')
# translation y-axis
ax1.set_ylim(0,28)

# add a text at point[2,25]in ax1 coordinate system
# ax1.text(2,25,'aaa')
    
#  traversal every bar
for rec in rect:
    x = rec.get_x()
    height = rec.get_height()
    ax1.text(x + 0.2,height*1.02,str(height)+'w')
```

### 4. Load web data and show them with matplotlib

```python
# Use request modle to load json string from api
import requests
import json
dataStr = requests.get('https://api.douban.com/v2/book/1220562').text
print('Origal data type:', type(dataStr))
dataDict = json.loads(dataStr)
print('After trans data type:', type(dataDict))
```

```python
tags = dataDict['tags']
# get all counts and name colnme form tags
x, y = [],[]
for tag in tags:
    x.append(tag['name'])
    y.append(tag['count'])
```

```python
# Use x and y to draw a bar chart
plt.bar(x, y, label='Hot words in search')
plt.xlabel('x-axis The words')
plt.ylabel('y-axis The rank of hot')
plt.title('Hot words in search')
plt.legend()
```
